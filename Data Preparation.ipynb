# **Supermarket Dataset**

Dari dataset yang dimiliki terdapat beberapa data yang tersedia:

---

**1. Order ID** – Kode unik untuk setiap transaksi pesanan.\
**2. Order Date** – Tanggal ketika pesanan dibuat.\
**3. Ship Date** – Tanggal ketika pesanan dikirim ke pelanggan.\
**4. Ship Mode** – Metode atau jenis layanan pengiriman yang digunakan.\
**5. Customer ID** – Kode unik untuk mengidentifikasi pelanggan.\
**6. Customer Name** – Nama pelanggan yang melakukan pembelian.\
**7. Segment** – Kategori segmen pelanggan, seperti konsumen, korporat, atau home office.\
**8. Country/Region** – Negara atau wilayah tempat pesanan dikirim.\
**9. City** – Kota tujuan pengiriman pesanan.\
**10. State/Province** – Provinsi atau negara bagian tujuan pengiriman.\
**11. Postal Code** – Kode pos lokasi pengiriman pesanan.\
**12. Region** – Wilayah geografis yang lebih luas tempat pengiriman dilakukan.\
**13. Product ID** – Kode unik untuk mengidentifikasi produk.\
**14. Category** – Kategori utama produk, seperti furniture, office supplies, atau technology.\
**15. Sub-Category** – Kategori lebih spesifik di dalam kategori utama produk.\
**16. Product Name** – Nama lengkap produk yang dibeli.\
**17. Sales** – Nilai penjualan total untuk produk tersebut dalam transaksi.\
**18. Quantity** – Jumlah unit produk yang dibeli.\
**19. Discount** – Persentase potongan harga yang diberikan pada produk.\
**20. Profit** – Laba bersih yang dihasilkan dari penjualan setelah dikurangi biaya.\

```python
# Salin dataframe untuk pembersihan data
df_clean = df.copy()
# Mengubah kolom Order Date ke format datetime
df_clean['Order Date'] = pd.to_datetime(df_clean['Order Date'], dayfirst=True)
# Mengubah kolom Ship Date ke format datetime
df_clean['Ship Date'] = pd.to_datetime(df_clean['Ship Date'], dayfirst=True)
# Menambahkan kolom Quartal
df_clean['Quarter'] = 'Q' + df_clean['Order Date'].dt.quarter.astype(str)
# Menambahkan kolom Delivery Time
df_clean['Delivery Time'] = (df_clean['Ship Date'] - df_clean['Order Date']).dt.days
# Menambahkan kolom Unit Price
df_clean['Unit Price'] = df_clean['Sales'] / (df_clean['Quantity']*(1-df_clean['Discount']))
# Menambahkan Kolom Discount Amount
df_clean['Discount Amount'] = df_clean['Unit Price'] * df_clean['Discount'] * df_clean['Quantity']
# Menghapus kolom yang tidak perlu (ï»¿Row ID)
df_clean.drop('ï»¿Row ID', axis=1, inplace=True)
# Menghapus kolom redundan (Postal Code)
df_clean.drop('Postal Code', axis=1, inplace=True)

df_clean.head().round(2)
```

**Membagi Transaksi ke dalam Beberapa Kategori**

```python
# Batas ekstrem Sales
sales_q3 = df_clean['Sales'].quantile(0.75)
sales_iqr = sales_q3 - df_clean['Sales'].quantile(0.25)
sales_extreme_threshold = sales_q3 + 1.5 * sales_iqr

# Batas ekstrem Discount
discount_q3 = df_clean['Discount'].quantile(0.75)
discount_iqr = discount_q3 - df_clean['Discount'].quantile(0.25)
discount_extreme_threshold = discount_q3 + 1.5 * discount_iqr
```

```python
def categorize_transaction(row):
    # Prioritas 1: Extreme Loss Discount
    if (row['Discount'] > discount_extreme_threshold) and (row['Profit'] < 0):
        return 'Extreme Loss Discount'

    # Prioritas 2: Operational Loss
    elif (row['Discount'] > 0) and (row['Profit'] < 0):
        return 'Operational Loss'

    # Prioritas 3: Premium Sales
    elif (row['Sales'] > sales_extreme_threshold) and (row['Profit'] >= 0):
        return 'Premium Sales'

    # Sisanya: Normal
    else:
        return 'Normal'

# Tambahkan kolom kategori ke df_clean
df_clean['Transaction Category'] = df_clean.apply(categorize_transaction, axis=1)
```
